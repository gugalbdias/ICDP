```python
import pandas as pd
import matplotlib.pyplot as plt

# Lê o arquivo base:
dados_df = pd.read_csv("https://raw.githubusercontent.com/drobaina/Python/main/microdados_ed_basica_RJ_2022.csv", header=0)

dados_df.head()
```

```python
dados_df = dados_df.drop(columns=["Unnamed: 0"], errors='ignore')

dados_df.head()
```

```python
#Filtrando os dados:
dados_df = dados_df[dados_df['NO_MUNICIPIO'] == 'Niterói']
dados_df = dados_df[dados_df['TP_SITUACAO_FUNCIONAMENTO'] == 1]

dados_df.head()
```

```python
#Selecionando as colunas
colunas_desejadas = [
    "NU_ANO_CENSO",
    "NO_MUNICIPIO",
    "TP_DEPENDENCIA",
    "IN_AGUA_POTAVEL",
    "IN_INF",
    "IN_MED"
]

dados_df = dados_df[colunas_desejadas]


dados_df.head()
```

```python
# Agrupando e contando o número de ocorrências em cada grupo
agrupado_df = dados_df.groupby(["NU_ANO_CENSO", "TP_DEPENDENCIA", "IN_AGUA_POTAVEL", "IN_INF", "IN_MED"]).size()
agrupado_df = agrupado_df.reset_index(name="NO_OBS")

# Mostrar o DataFrame resultante
print(agrupado_df)
```

```python
### total de escolas:
obs_totais = agrupado_df['NO_OBS'].sum()

### total de escolas que oferecem água potável:
filtro_ap = agrupado_df['IN_AGUA_POTAVEL']==1
obs_ap = agrupado_df[filtro_ap]['NO_OBS'].sum()

# calcular a porcentagem de escolas que oferecem água potável:
porc_ap = (obs_ap / obs_totais) * 100

plt.figure(figsize=(10,6))
plt.pie(porc_ap, labels=agrupado_df['TP_DEPENDENCIA'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors) #  escala de cores plt.cm.Paired.colors
plt.title("Distribuição das observações por tipo de dependência")
plt.show()
```
